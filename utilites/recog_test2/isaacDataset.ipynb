```python
import pandas as pd
path = r"E:\OneDrive\_workflow\binding of issac FLASH item description\recog_test2\dataset"
path = r".\dataset"
df_list = (pd.read_csv(f) for f in [path+r'\1.csv', path+r'.\2.csv', path+r'.\3.csv', path+r'.\4.csv'])
combined_df = pd.concat(df_list, ignore_index=True)
combined_df = combined_df.rename(columns={"isReload": "reloadCnt", "descr":"descrRU"})
combined_df.head()
display(combined_df.head())
display(combined_df.describe())
```

```python
combined_df['imgName'] = combined_df.apply(lambda x: x['imgName'].replace('.png','.webp'), axis = 1)    
combined_df['isTrinket'] = False
#combined_df["urlToPage"] = ''
combined_df["imgSrc_dataset"] = ".\dataset\itemsImgs" + '\\' + combined_df["imgName"]
combined_df.head()
```

```python
trinkets_df = pd.read_csv(path+r'\5.csv')
#combined_df = pd.concat([combined_df, trinkets_df], ignore_index=True)
#display(trinkets_df.head())
#display(trinkets_df.describe())
trinkets_df["imgSrc_dataset"] = ".\dataset\itemsImgs" + '\\' + trinkets_df["imgName"]
trinkets_df = trinkets_df.rename(columns={"descr":"descrRU"})
#pd.set_option('display.max_columns', 1500)
trinkets_df = trinkets_df.drop('urlToPage', axis=1)
display(trinkets_df[["nameEN","imgName","imgSrc","descrRU"]])
```
